## **Data Visualization**

**Data visulization using titanic dataset**

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams["figure.figsize"] = [10,5]
# Ignore warnings

import warnings
# Set the warning filter to ignore FutureWarning
warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python
full_data = pd.read_csv('/content/housing.csv')
```

```python
full_data.head()
```

```python
full_data.shape
```

**Histplot**

```python

sns.histplot(full_data['housing_median_age'],kde=False)
plt.show()
```

**Distplot**

```python

plt.figure(figsize=(8,8))
sns.distplot(full_data['housing_median_age'])
plt.show()
```

**lineplot**

```python

plt.figure(figsize=(8,8))
sns.lineplot(x="housing_median_age",y="population", data=full_data)
plt.show()
```

**Barplot**

```python

plt.figure(figsize=(10,8))
sns.barplot(x="ocean_proximity", y="housing_median_age",hue="housing_median_age",data=full_data)
plt.show()
```
